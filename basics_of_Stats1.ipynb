```python
# Boxplot
sns.boxplot(DF['Volume'])
plt.xlabel("Volume")
plt.ylabel("")
plt.title("Boxplot")
# Identification of Outliers using IQR method.
Q1 = DF['Volume'].quantile(0.25)
Q3 = DF['Volume'].quantile(0.75)
IQR = Q3-Q1
print(f"First Quantile Q1 is: {Q1}")
print(f"Third Quantile Q3 is: {Q3}") 
print(f"Interquantile Range is: {IQR}")
LE = Q1-1.5*IQR
UE = Q3-1.5*IQR
outliers=DF[(DF['Volume']<LE) | (DF['Volume']>UE)]
print(f'outliers:{outliers}')
```

```python
'''Standardization of Numerical Variables
●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.
●	Steps:
	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma
	Show before and after comparisons of the data distributions.'''
# When a dataset contains different magnitudes of Data, Feature scaling methods like Standardization is used. 
# After Standardization all the numerical values will be converted to values ranging from -3 to +3.
DF.head() #dataset before standardization.
```

```python
def StandardScaler(X):
    return (X-X.mean())/X.std()
StandardScaler(DF) #dataset after Standardization.
```
