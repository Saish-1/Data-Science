```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_norm = scaler.fit_transform(df)
```

```python
df_norm
```

```python
df_norm = pd.DataFrame(df_norm)
```

```python
#df1.columns = ['ID#','Balance','Qual_miles','cc1_miles','cc2_miles','cc3_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','Flight_trans_12','Days_since_enroll','Award?']
df_norm.columns = df.columns
df_norm
```

```python
df1 = df_norm.copy() # creating a new dataframe so that the normalised dataframe is saved for future uses.
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3).fit(df1)
cluster_names = model.predict(df1)
```

```python
cluster_names
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df_norm,cluster_names)
```

```python
model.inertia_
```

```python
WCSS = []
for i in range(1,11):
    model = KMeans(n_clusters = i).fit(df_norm)
    WCSS.append(model.inertia_)
```

```python
plt.title('Elbow Method')
plt.plot(range(1,11),WCSS)
figure = plt.figure(figsize=(8,6))
plt.show()
```

```python
# number of clusters can be changed to 4-6 clusters.
```

```python
model = KMeans(n_clusters = 5)
clusters = model.fit_predict(df_norm)
clusters
```

```python
df1['cluster_name'] = clusters
```

```python
df1
```

```python
model.inertia_
```

```python
df1.plot("Days_since_enroll","Qual_miles", c = model.labels_,kind='scatter',s=50, cmap ='plasma',edgecolor='black')
```

```python
silhouette_score(df1,clusters)
```

```python
# Heirarchical Clustering
from sklearn.cluster import AgglomerativeClustering
hc_model = AgglomerativeClustering(n_clusters = 4 , linkage = 'average' , metric = 'euclidean')
```

```python
df2 = df_norm.copy()
```

```python
Clusters = hc_model.fit_predict(df2)
Clusters
```

```python
import scipy.cluster.hierarchy as sch #for dendrogram
dendogram = sch.dendrogram(sch.linkage(df2,method='average'),labels=df2.index)
figure = plt.figure(figsize = (20,5))
```
