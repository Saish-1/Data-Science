As the data given was clumsy I used a delimiter with coma(,) to seperate all items and gave header = None to give index column wise.

```python
import pandas as pd
df = pd.read_excel("Online retail copy.xlsx",header=None)
df
```

```python
data = pd.get_dummies(df,dtype='int')
data
```

```python
import warnings
warnings.filterwarnings('ignore') #to ignore warnings
from mlxtend.frequent_patterns import apriori, association_rules
frequent_items = apriori(data, min_support = 0.001, use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items,metric = 'lift',min_threshold=1)
rules
```
